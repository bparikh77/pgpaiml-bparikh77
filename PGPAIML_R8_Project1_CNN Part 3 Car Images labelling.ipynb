# Import necessary libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Read the data from the previously prepared excel file

```python
car_data_df = pd.read_excel('Car Images.xlsx')
```

```python
car_data_df
```

# Merge the model and colour to prepare a label. Prepare a list of images and labels

```python
car_images = []
car_labels = []
image_name = ''
path = 'C:/Users/Bhavik Parikh/Python/Computer Vision/Project'
os.chdir(path)
for i in range(0,len(car_data_df)):
    image_name = car_data_df['Image Name'][i]
    img = plt.imread(path + '/Data/' + image_name)
    car_images.append(img)
    car_labels.append(car_data_df['Make'][i] + ' ' + car_data_df['Colour'][i])
```

# Display the list of images and labels using matplotlib

```python
def plot_images(images, labels, num_row, num_col):
   
    fig, axes = plt.subplots(num_row, num_col, figsize=(5*num_col,3*num_row))
    for i in range(0, len(images)):
        ax = axes[i//num_col, i%num_col]       
        ax.imshow(images[i])
        ax.set_title(labels[i], weight='bold', fontsize=20)
    plt.tight_layout()    
    plt.show()

plot_images(car_images, car_labels , num_row=3, num_col=5)
```
